# 🎮 Tarea 2 — API REST: RAWG Video Games Database

Este notebook consume la API de RAWG para extraer, analizar y comparar datos de videojuegos.

Documentación oficial: https://rawg.io/apidocs

```python
import requests
import pandas as pd
import os

API_KEY = "TU_API_KEY_AQUI"  # Reemplazar con tu key local
BASE_URL = "https://api.rawg.io/api"

class RAWGClient:
    def __init__(self, api_key):
        self.api_key = api_key
        self.total_requests = 0

    def get(self, endpoint, params={}):
        params["key"] = self.api_key
        url = f"{BASE_URL}/{endpoint}"
        response = requests.get(url, params=params)
        self.total_requests += 1
        response.raise_for_status()
        return response.json()

    def resumen_requests(self):
        print(f"📊 Total de requests realizadas: {self.total_requests}")

client = RAWGClient(API_KEY)
print("✅ Cliente RAWG inicializado correctamente")
```

## 🟢 Part A — Exploración General

### A1 — Total de juegos registrados en RAWG

```python
# Consultamos el endpoint /games y extraemos el campo 'count'
data = client.get("games")
total_games = data["count"]
print(f"🎮 RAWG tiene un total de {total_games:,} juegos registrados.")
```

## 🔵 Part B — Análisis por Categoría

### B1 — Top 5 juegos mejor puntuados según Metacritic

```python
# Consultamos los juegos ordenados por metacritic score de mayor a menor
data = client.get("games", params={
    "ordering": "-metacritic",
    "page_size": 5
})

print("🏆 Top 5 juegos con mayor puntuación Metacritic:\n")
for i, game in enumerate(data["results"], 1):
    print(f"{i}. {game['name']}")
    print(f"   ⭐ Rating: {game['rating']} | 🎯 Metacritic: {game['metacritic']}")
    print()
```

### B2 — Top 10 mejores juegos en Steam (store_id=1)

```python
# Consultamos los juegos disponibles en Steam ordenados por rating
data = client.get("games", params={
    "stores": 1,
    "ordering": "-rating",
    "page_size": 10
})

print("🎮 Top 10 mejores juegos en Steam:\n")
for i, game in enumerate(data["results"], 1):
    print(f"{i}. {game['name']}")
    print(f"   ⭐ Rating: {game['rating']} | 🎯 Metacritic: {game['metacritic']}")
    print()
```

## 🟡 Part C — Comparaciones

### C1 — Top 5 PC (platform_id=4) vs Top 5 PS5 (platform_id=187)

```python
# Top 5 juegos en PC
pc_data = client.get("games", params={
    "platforms": 4,
    "ordering": "-rating",
    "page_size": 5
})

# Top 5 juegos en PS5
ps5_data = client.get("games", params={
    "platforms": 187,
    "ordering": "-rating",
    "page_size": 5
})

# Mostrar resultados
print("🖥️ Top 5 juegos en PC:\n")
pc_ratings = []
for i, game in enumerate(pc_data["results"], 1):
    print(f"{i}. {game['name']} | ⭐ {game['rating']}")
    pc_ratings.append(game["rating"])

print(f"\n📊 Rating promedio PC: {sum(pc_ratings)/len(pc_ratings):.2f}")

print("\n🎮 Top 5 juegos en PS5:\n")
ps5_ratings = []
for i, game in enumerate(ps5_data["results"], 1):
    print(f"{i}. {game['name']} | ⭐ {game['rating']}")
    ps5_ratings.append(game["rating"])

print(f"\n📊 Rating promedio PS5: {sum(ps5_ratings)/len(ps5_ratings):.2f}")

# Conclusión
if sum(pc_ratings)/len(pc_ratings) > sum(ps5_ratings)/len(ps5_ratings):
    print("\n✅ PC tiene los juegos mejor puntuados.")
else:
    print("\n✅ PS5 tiene los juegos mejor puntuados.")
```

### C2 — Tabla comparativa de 3 juegos famosos

```python
# Buscamos 3 juegos famosos por nombre
juegos_famosos = ["Red Dead Redemption 2", "God of War", "Minecraft"]
resultados = []

for juego in juegos_famosos:
    data = client.get("games", params={"search": juego, "page_size": 1})
    if data["results"]:
        g = data["results"][0]
        # Extraer géneros y plataformas
        generos = ", ".join([genre["name"] for genre in g.get("genres", [])])
        plataformas = ", ".join([p["platform"]["name"] for p in g.get("platforms", [])])
        resultados.append({
            "name": g["name"],
            "rating": g["rating"],
            "metacritic": g["metacritic"],
            "genres": generos,
            "platforms": plataformas
        })

# Mostrar como tabla
df_comparacion = pd.DataFrame(resultados)
print("🎮 Tabla comparativa de 3 juegos famosos:\n")
print(df_comparacion.to_string(index=False))
```

### C3 — Comparación de géneros: ¿cuál produce los mejores juegos?

```python
# Consultamos top 5 juegos de 4 géneros diferentes
generos = {
    "Action": 4,
    "RPG": 5,
    "Shooter": 2,
    "Strategy": 10
}

resultados_generos = []

for nombre_genero, genre_id in generos.items():
    data = client.get("games", params={
        "genres": genre_id,
        "ordering": "-rating",
        "page_size": 5
    })
    ratings = [g["rating"] for g in data["results"]]
    promedio = sum(ratings) / len(ratings)
    resultados_generos.append({
        "genre": nombre_genero,
        "avg_rating": round(promedio, 2),
        "top_game": data["results"][0]["name"]
    })
    print(f"🎮 {nombre_genero}: promedio {promedio:.2f} | Mejor juego: {data['results'][0]['name']}")

# Determinar el mejor género
mejor = max(resultados_generos, key=lambda x: x["avg_rating"])
print(f"\n🏆 El género con mejor rating promedio es: {mejor['genre']} ({mejor['avg_rating']})")
```

### C4 — Comparación de juegos por año: ¿en qué año salieron los mejores?

```python
# Comparamos los mejores juegos de 3 años diferentes
years = [2015, 2020, 2023]
resultados_años = []

for year in years:
    data = client.get("games", params={
        "dates": f"{year}-01-01,{year}-12-31",
        "ordering": "-metacritic",
        "page_size": 5
    })
    
    metacritic_scores = [g["metacritic"] for g in data["results"] if g["metacritic"]]
    promedio = sum(metacritic_scores) / len(metacritic_scores) if metacritic_scores else 0
    resultados_años.append({
        "year": year,
        "avg_metacritic": round(promedio, 2),
        "top_game": data["results"][0]["name"]
    })
    print(f"📅 {year}: promedio metacritic {promedio:.2f} | Mejor juego: {data['results'][0]['name']}")

# Determinar el mejor año
mejor_año = max(resultados_años, key=lambda x: x["avg_metacritic"])
print(f"\n🏆 El año con mayor metacritic promedio fue: {mejor_año['year']} ({mejor_año['avg_metacritic']})")
```

### C5 — Top 20 juegos de todos los tiempos exportado a CSV

```python
# Consultamos el top 20 juegos ordenados por rating
data = client.get("games", params={
    "ordering": "-rating",
    "page_size": 20
})

# Construimos el dataframe con las columnas requeridas
top20 = []
for g in data["results"]:
    main_genre = g["genres"][0]["name"] if g.get("genres") else "N/A"
    top20.append({
        "name": g["name"],
        "rating": g["rating"],
        "metacritic": g["metacritic"],
        "release_date": g["released"],
        "main_genre": main_genre
    })

df_top20 = pd.DataFrame(top20)

# Exportar a CSV
os.makedirs("output", exist_ok=True)
df_top20.to_csv("output/top20_rawg.csv", index=False, encoding="utf-8-sig")

print("✅ Archivo top20_rawg.csv exportado correctamente.\n")
print("📋 Primeras 5 filas:")
df_top20.head()
```
